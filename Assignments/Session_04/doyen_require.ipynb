###  1- How many distinct account_id on each platform?

 ### 2- How many distinct ad_id on each platform?

```python
sp.groupby('platform').ad_id.nunique()
```

### 3- For each category_name how many ad is advertised?

```python
sp.groupby('category_name').ad_id.nunique()
```

### 4- Give me top 5 category_name having the highest number of ads?


```python
sp.groupby('category_name').ad_id.nunique().nlargest(3)
```

###  5- Give me top 5 category_name having lowest number of ads?

```python
sp.groupby('category_name').ad_id.nunique().sort_values(ascending=False).head(5)
```

### 6- Give me the average price for all distinct ads within each category?

```python
sp.groupby(['category_name','ad_id']).price.mean()
```

```python
sp.groupby(['category_name','ad_id']).agg(prices_each_id=('price','mean'))
```

### 7- Give me the average price for all distinct ads within each region_name?

```python
sp.groupby(['region_name','ad_id']).agg(prices_each_ads=("price",'sum'))
```

### Give details of condition_ad_name for all ads within each category? For example, for category cars, how may ads is 'Moi', 'Da su dung', and 'NaN'

```python
sp.groupby(['category_name','condition_ad_name']).ad_id.count()
```

```python
sp.groupby(['category_name','condition_ad_name']).agg(number_of_ads=('ad_id','count'))
```
