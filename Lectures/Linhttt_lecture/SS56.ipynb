```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sin function')
plt.xlabel('X values')
plt.ylabel('Sin of X')
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
cos = np.cos(x)

plt.plot(x, sin, label ='sin', c='red', linestyle='-.') 
plt.plot(x,cos, label ='cos')
plt.title('Funtionals plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
ax =plt.axes()
ax.plot(x,sin)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,sin)
plt.title('Sin func')

plt.subplot(1,2,2)
plt.plot(x,cos,c='r')
plt.title('Cos func')

plt.show()
```

```python
plt.figure(figsize=(15,5))

ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

ax1.plot(x,sin)
ax2.plot(x,cos,c='r')

ax1.set_title('Sin func')
ax2.set_title('Cos func')

plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos,c='r')
plt.show()
```
