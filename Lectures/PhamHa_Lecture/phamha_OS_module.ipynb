```python
#Code 
import sys
from socket import *

# port_scan.py <host> <start_port>-<end_port>
host = sys.argv[1]
portstrs = sys.argv[2].split('-')
start_port = int(portstrs[0])
end_port = int(portstrs[1])

target_ip = gethostbyname(host)
opened_ports = []

for port in range(start_port, end_port):
    sock = socket(AF_INET, SOCK_STREAM)
    sock.settimeout(10)
    result = sock.connect_ex((target_ip, port))
    if result == 0:
        opened_ports.append(port)

print("Opened ports:")

for i in opened_ports:
    print(i)
```

```python
def get_open_port():
        import socket
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.bind(("",0))
        s.listen(1)
        port = s.getsockname()[1]
        s.close()
        return port
print('Cac port dang mo:',get_open_port())
```

```python
import socket
def isOpen(ip,port):
   s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
   try:
      s.connect((ip, int(port)))
      s.shutdown(2)
      return True
   except:
      return False
i ='local'
for j in range(70,90):
    if isOpen(i,j):
        print('IP:',i,'Port:',j)
```

```python
# Method 1: Lấy địa chỉ MAC của máy tính: Hạn chế cách này là địa chỉ MAC không giống định dạng
# Python Program to compute
# MAC address of host
# using UUID module
  
import uuid
  
# printing the value of unique MAC
# address using uuid and getnode() function 
print (hex(uuid.getnode()))
```

```python
# Method 2:Lấy địa chỉ MAC của máy tính: Hạn chế cách này là code có vẻ phức tạp, khó hiểu
# Python 3 code to print MAC
# in formatted way.
  
import uuid
  
# joins elements of getnode() after each 2 digits.
  
mac=':'.join(['{:02x}'.format((uuid.getnode() >> ele) & 0xff)
for ele in range(0,8*6,8)][::-1])

print ("The MAC address in formatted way is : ", mac)
```

```python
# Method 3: Lấy địa chỉ MAC của máy tính: 
# Python 3 code to print MAC
# in formatted way and easier
# to understand
  
import re, uuid
  
# joins elements of getnode() after each 2 digits.
# using regex expression
mac3=':'.join(re.findall('..', '%012x' % uuid.getnode()))
print ("The MAC address in formatted and less complex way is : ", mac3)
```
