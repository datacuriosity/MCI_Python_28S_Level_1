```python
#tính trung bình chi tiêu theo nam hay nữ
tip_df.groupby('sex').total_bill.mean()
```

```python
#tính trung bình chi tiêu và tiền tip theo nam hay nữ
tip_df.groupby('sex').agg({'total_bill':'mean','tip':['mean','max','count']})
```

```python
#tính trung bình chi tiêu và tiền tip theo nam hay nữ
tip_df.groupby(['sex','smoker']).agg({'total_bill':'mean','tip':['mean','max','count']})
```

```python
#tính tổng các trường kiểu số theo giới tính và hút thuốc
tip_df.groupby(['sex','smoker']).sum()
```

### 5.4 Join / Merge:

![](https://xuanthulab.net/photo/sql-joins-4213.jpg)

```python
left=tip_df.iloc[:10,:2]
right=tip_df.iloc[5:15,2:4]
```

```python
left
```

```python
right
```

```python
# Merge left với right
left.merge(right,how='inner',left_index=True,right_index=True)
```

```python
left.merge(right,how='outer',left_index=True,right_index=True)
```

```python
left.merge(right,how='left',left_index=True,right_index=True)
```

```python
left.merge(right,how='right',left_index=True,right_index=True)
```
