```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sin function')
plt.xlabel('X values')
plt.ylabel('Sin of X')
plt.grid()
plt.show()
```

```python
x=np.linspace(-10,10,100)
sin=np.sin(x)
cos=np.cos(x)

plt.plot(x,sin,label='SIN',color='red' ,linestyle='-.')
plt.plot(x,cos,label='COS',color='green',linestyle='-')
plt.title('Functions plot')
plt.xlabel('X value')
plt.ylabel('Sin of X')
plt.legend(loc='best')
plt.show()
```

```python
ax=plt.axes()
ax.plot(x,sin)
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('Sin function')
ax.legend(loc='upper left')
plt.show()
```

```python
# Vẽ nhiều đồ thị trên cùng figure
plt.figure(figsize=(15,5)) #muốn tăng kích thước đồ thị phải tăng kích thước figuure
plt.subplot(1,2,1) #1 hàng, 2 cột
plt.plot(x,sin)
plt.title('Sin func')

plt.subplot(1,2,2)
plt.plot(x,cos,c='r')
plt.title('Cos func')


plt.show()
```

```python
# Dùng biến cho phép sử dụng lại nhiều lần sau đó
plt.figure(figsize=(15,5)) #muốn tăng kích thước đồ thị phải tăng kích thước figuure
ax_1=plt.subplot(1,2,1) 
ax_2=plt.subplot(1,2,2) 

ax_1.plot(x,sin)
ax_2.plot(x,cos,c='r')

ax_1.set_title('Sin func')
ax_2.set_title('Cos func')

plt.show()
```

```python
# Cách khác
fig,ax=plt.subplots(1,2,figsize=(15,5)) 
ax[0].plot(x,sin)
ax[1].plot(x,cos,color='red')
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
tips.groupby('sex').total_bill.sum()
```

```python
series=tips.groupby('sex').total_bill.sum()
plt.pie(series,labels=series.index,autopct='%.2f%%')
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
series=tips.groupby('sex').total_bill.sum()
ax[0].pie(series,labels=series.index,autopct='%.2f%%')
ax[1].pie(tips.groupby('day').total_bill.sum().values,
                      labels=tips.groupby('day').total_bill.sum().index,
                      autopct='%.2f%%',explode=[0,0.1,0,0])
```
