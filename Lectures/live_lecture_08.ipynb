```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

- Figure: chứa nhiều axes
- Axes: đồ thị

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sin function')
plt.xlabel('X values')
plt.ylabel('Sin of X')
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
sin = np.sin(x)
cos = np.cos(x)

plt.plot(x, sin, label='Sin', color='red', linestyle='-.')
plt.plot(x, cos, label='Cos', linestyle='--')
plt.title('Functions plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend(loc='best')
plt.show()
```

```python
ax = plt.axes()
ax.plot(x, sin, label='Sin')
ax.plot(x, cos, label='Cos')
ax.set(xlabel='X label', ylabel='Y label', title='Function')
# ax.set_xlabel('X label')
# ax.set_ylabel('Y label')
# ax.set_title('Sin function')
ax.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(x, sin)
plt.title('Sin func')

plt.subplot(1, 2, 2)
plt.plot(x, cos, c='r')
plt.title('Cos func')

plt.show()
```

```python
plt.figure(figsize=(15, 5))
ax_1 = plt.subplot(1, 2, 1)
ax_2 = plt.subplot(1, 2, 2)

ax_1.plot(x, sin)
ax_2.plot(x, cos, c='r')

ax_1.set_title('Sin func')
ax_2.set_title('Cos func')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(x, sin)
ax[1].plot(x, cos)
ax[0].set(xlabel='', ylabel='', title='Sin function')
ax[1].set(xlabel='', ylabel='', title='Cos function')
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(2)
```

```python
tips.groupby('sex').total_bill.sum()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

series = tips.groupby('sex').total_bill.sum()
ax[0].pie(series, labels=series.index, autopct='%.2f%%');
ax[1].pie(tips.groupby('day').total_bill.sum().values, 
          labels=tips.groupby('day').total_bill.sum().index, 
          autopct='%.2f%%', explode=[0, 0.2, 0.0, 0]);
```

```python
# fig, ax = plt.subplots(1, 2, figsize=(25, 7))
plt.figure(figsize=(7, 7))
tips.query("sex == 'Male'").groupby('day').total_bill.sum().plot(kind='pie', autopct='%.2f%%', figsize=(8, 8), 
                                                                 explode=[0, 0.2, 0, 0]);

# tips.query("sex == 'Male'").groupby('day').total_bill.sum().plot(kind='bar', ax=ax[1]);
```
