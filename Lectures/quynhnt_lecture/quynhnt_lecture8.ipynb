```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.title('Sin function')
plt.xlabel('X values')
plt.ylabel('Sin of X')
plt.show()
```

```python
x=np.linspace(-10,10,100)
sin=np.sin(x)
cos=np.cos(x)

plt.plot(x,sin,label='Sin',color='red')
plt.plot(x,cos,label='Cos',linestyle='--')

plt.title('Functions plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend(loc='best')

plt.show()
```

```python
ax = plt.axes()
ax.plot(x,sin,label='sin')
ax.plot(y,sin,label='cos')

ax.set(xlabel='X label', ylabel='Y label', )
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(x,sin)
plt.title('Sin func')

plt.subplot(1,2,2)
plt.plot(x,cos,color='red')
plt.title('Cos func')

plt.show()
```

```python
plt.figure(figsize=(15,5))

ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

ax1.plot(x,sin)
ax2.plot(x,cos,color='red')

ax1.set_title('Sin func')
ax2.set_title('Cos func')

plt.show()
```

```python
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].plot(x,sin)
ax[1].plot(x,cos)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

```python
sr=tips.groupby('sex').total_bill.sum()
sr
```

```python
plt.pie(sr,autopct='%.2f%%',labels=sr.index);
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))

sr1=tips.groupby('sex').total_bill.sum()
sr2=tips.groupby('day').total_bill.sum()

ax[0].pie(sr1,autopct='%.2f%%',labels=sr1.index);
ax[1].pie(sr2,autopct='%.2f%%',labels=sr2.index, explode=[0,0.1,0,0]);
```
