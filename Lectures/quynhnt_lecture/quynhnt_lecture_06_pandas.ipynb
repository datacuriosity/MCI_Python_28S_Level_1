# 4. Operations

## 4.1 Group by

```python
tip_df
```

```python
tip_df.groupby('sex').total_bill.mean()
```

```python
tip_df['total']=tip_df['total_bill']+tip_df['tip']
tip_df.head(2)
```

```python
tip_df.groupby('sex')['total'].mean()
```

```python
tip_df.groupby('sex').agg({'total_bill':'mean','tip':
                          'mean'})
```

```python
tip_df.groupby(['sex','smoker'])[['total_bill','tip']].mean()
```

```python
tip_df.groupby(['sex','smoker']).mean()
```

```python
tip_df.groupby(['sex','smoker']).agg({'total_bill':('mean','max','median'),'tip':('mean','max','median')})
```

## 4.2 Pivot
