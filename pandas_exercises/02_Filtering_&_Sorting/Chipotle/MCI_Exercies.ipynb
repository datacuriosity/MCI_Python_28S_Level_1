### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', delimiter='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo.head(5)
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo.item_price.apply(lambda x : float(x.replace('$', '')))
```

```python
chipo.query("(item_price / quantity) > 10").item_name.nunique()
```

```python
chipo.query("(item_price / quantity) > 10").item_name.unique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo['price_per_item'] = chipo.item_price / chipo.quantity
chipo.query("item_name == 'Chicken Bowl'") # dynamic pricing
```

```python
chipo.groupby('item_name').price_per_item.mean()
```

```python
pd.options.display.max_rows = 2000
```

```python
chipo.groupby(['item_name', 'choice_description']).price_per_item.mean().to_frame()
```

### Step 6. Sort by the name of the item

```python
# chipo.item_name.sort_values()
```

```python
chipo.sort_values(by=['item_name', 'item_price', 'quantity'], ascending=[True, False, False], ignore_index=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo
```

```python
chipo.query("price_per_item == price_per_item.max()", engine='python').groupby('item_name').quantity.sum()
```

```python
chipo[chipo.price_per_item == chipo.price_per_item.max()].groupby('item_name').quantity.sum()
```

```python
ax = chipo[chipo.price_per_item == chipo.price_per_item.max()].groupby('item_name').quantity.sum().plot(kind='bar', 
                                title='Quantity of most expensive items', xlabel='Products', ylabel='Quantity', 
                                color='purple');
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right');
```
