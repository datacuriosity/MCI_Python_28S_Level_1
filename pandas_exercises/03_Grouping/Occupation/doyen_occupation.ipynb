### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
oc=pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',delimiter='|')
oc
```

### Step 4. Discover what is the mean age per occupation

```python
oc.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
oc_1=oc.groupby(['occupation']).gender.agg(['count'])
oc_2=oc.groupby(['occupation','gender']).gender.agg(['count']).query("gender=='M'")
```

```python
oc_4=oc_2.div(oc_1, level='occupation')
oc_4.sort_values(by='count')
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
oc.groupby('occupation').age.agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
oc.groupby(['occupation','gender']).age.agg(['min','max'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
oc_1=oc.groupby(['occupation']).gender.agg(['count'])
oc_5=oc.groupby(['occupation','gender']).gender.agg(['count'])
```

```python
oc_3=oc_5.div(oc_1, level='occupation')*100
oc_3
```
